```python
import pandas as pd
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from wordcloud import WordCloud, STOPWORDS
from IPython.display import Image
import matplotlib.pyplot as plt
chromedriver_path = "C:/Users/deera/anaconda3/pkgs/chromedriver"
search = input("Enter topic:")
driver = webdriver.Chrome(chromedriver_path)
url = f"https://en.wikipedia.org/wiki/{search}"
driver.get(url)
page_source = driver.page_source
soup = BeautifulSoup(page_source, 'lxml')
topic = soup.find("h1",class_="firstHeading")
print(topic.text)
full=""
for content in soup.find_all("div",class_="mw-parser-output"):
    for k in content.find_all("p"):
        full = k.text + full
    
text = full
stopwords = STOPWORDS
wc = WordCloud(
    background_color = 'white',
    stopwords = stopwords,
    height = 600,
    width = 400,       
)
wc.generate(text)
plt.figure(figsize=(10,10))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
